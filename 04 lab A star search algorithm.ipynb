```python
# Rohan Sanjay Pol
# Roll No- M240006CS
# A* Search Algorithm

def a_star(graph, heuristic, start, goal):
    open_list = {start: (0 + heuristic[start], 0)}  
    closed_list = {}
    parent_map = {}  
    while open_list:
        current_node = min(open_list, key=lambda node: open_list[node][0])
        current_f_cost, current_g_cost = open_list[current_node]

        if current_node == goal:
            path = []
            while current_node in parent_map:
                path.append(current_node)
                current_node = parent_map[current_node]
            path.append(start)
            return path[::-1], current_f_cost  

        closed_list[current_node] = open_list.pop(current_node)

        for neighbor, cost in graph[current_node]:
            if neighbor in closed_list:
                continue

            new_g_cost = current_g_cost + cost
            new_f_cost = new_g_cost + heuristic[neighbor]

            if neighbor not in open_list or new_g_cost < open_list[neighbor][1]:
                open_list[neighbor] = (new_f_cost, new_g_cost)
                parent_map[neighbor] = current_node

    return None, float('inf')  

  graph = {
    'S': [('B',4), ('C',3)],
    'B': [('F', 5), ('E', 12)],
    'C': [('D', 7),('E',10)],
    'D': [('E', 2)],
    'E': [('G', 5)],
    'F': [('G',16)],
    'G': [],
}
heuristic ={
    'S': 14,
    'B': 12,
    'C': 11,
    'D': 6,
    'E': 4,
    'F': 11,
    'G': 0
}
start = 'S'
goal = 'G'

path, total_cost = a_star(graph, heuristic, start, goal)
print("Path:", path)
print("Total cost:", total_cost)
```
