```python
from queue import PriorityQueue
import copy
```

```python
def printBoard(board):
    for row in board:
        print(row)
    print()
```

```python
def calcAttackPos( chessBoard, row, col):
    n = len(chessBoard)
    attack =0

    # check horizontally
    i = row
    j = col
    j = j-1
    while j>=0:
        if chessBoard[i][j] == 'Q':
            attack += 1
        j = j-1
    
    # check upper diagonal
    i = row
    j = col
    j = j-1
    i = i-1
    while i>=0 and j>=0:
        if chessBoard[i][j] == 'Q':
            attack += 1
        j = j-1
        i = i-1
    
    # check lower diagonal
    i = row
    j = col
    j = j-1
    i = i+1
    while i<n and j>=0:
        if chessBoard[i][j] == 'Q':
            attack += 1
        j = j-1
        i = i+1

    return attack
```

```python
def heuristic(state):
    
    attack = 0

    n = len(state)
    for i in range(n):
        for j in range(n):
            if( state[i][j] == 'Q'):
                attack += calcAttackPos(state,i,j)

    return attack
```

```python
def hillClimb( start):

    n = len(start)
    state = copy.deepcopy(start)

    while True:

        print(heuristic(state))
        printBoard(state)

        if( heuristic(state) == 0):
            return

        pq = PriorityQueue()

        for i in range(n):
            for j in range(n):

                if state[i][j] == 'Q':

                    for k in range(n):
                        if i != k:
                            new_state = copy.deepcopy(state)
                            new_state[i][j] = ' '
                            new_state[k][j] = 'Q'
                            hVal = heuristic(new_state)
                            pq.put((hVal,new_state))
        
        hus, sts = pq.get()
        state = sts
    
```
