```python
def is_goal(state, goal_state):
    return state == goal_state

def get_neighbors(state):
    neighbors = []
    zero_index = state.index(0)
    row, col = zero_index // 3, zero_index % 3
    
    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]
    
    for direction in directions:
        new_row, new_col = row + direction[0], col + direction[1]
        if 0 <= new_row < 3 and 0 <= new_col < 3:
            new_index = new_row * 3 + new_col
            new_state = state[:]
            new_state[zero_index], new_state[new_index] = new_state[new_index], new_state[zero_index]
            neighbors.append(new_state)
    
    return neighbors

def dfs_limited(initial_state, goal_state, max_depth):
    stack = [(initial_state, [], 0)]  # Stack with (state, path, depth)
    visited = set()
    
    while stack:
        state, path, depth = stack.pop()
        
        if is_goal(state, goal_state):
            return path + [state]  # Return solution path
        
        if tuple(state) not in visited and depth < max_depth:
            visited.add(tuple(state))
            
            for neighbor in get_neighbors(state):
                if tuple(neighbor) not in visited:
                    stack.append((neighbor, path + [state], depth + 1))
    
    return None  # Return None if no solution is found

# Helper function to print the puzzle state
def print_puzzle(state):
    for i in range(0, 9, 3):
        print(state[i:i+3])
    print()

# Example usage
initial_state = [1, 2, 3, 4, 0, 5, 6, 7, 8] 
goal_state = [1, 2, 3, 4, 5, 6, 7, 8, 0]    
max_depth = 20  

solution = dfs_limited(initial_state, goal_state, max_depth)

if solution:
    print("Solution found!")
    for step in solution:
        print_puzzle(step)
else:
    print("No solution found within depth limit.")
```
