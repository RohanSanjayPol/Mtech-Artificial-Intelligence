# BFS 

```python
def bfs(graph, start):
    visited = []         
    queue = [start]       

    while queue:          
        node = queue.pop(0)  

        if node not in visited:
            print(node, end=" ")  
            visited.append(node)  

            for neighbor in graph[node]:
                if neighbor not in visited:
                    queue.append(neighbor)

graph = {
    '1': ['2', '3'],
    '2': ['4'],
    '3': ['4','7'],
    '4': ['5','6'],
    '5': [],
    '6': [],
    '7':[]
}

bfs(graph, '3') 
```

# DFS

```python
def dfs(graph, node, visited=None):
    if visited is None:
        visited = []  

    if node not in visited:
        print(node, end=" ")  
        visited.append(node)

        for neighbor in graph[node]:
            dfs(graph, neighbor, visited)

graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': [],
    'E': [],
    'F': [],
    'G': []
}

dfs(graph, 'C')  
```
