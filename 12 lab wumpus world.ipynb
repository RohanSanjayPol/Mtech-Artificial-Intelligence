```python
environment=[
    ["0","B","P","0"],
    ["S","0","B","0"],
    ["W","G","P","0"],
    ["S","0","0","0"]
    ]

Agent="@"
for i in environment:
    print(i)
```

```python
def AgentPosition(environment,Agent):
    environment[0][0]=Agent
    for i in range(len(environment)):
        for j in range(len(environment[i])):
          print(environment[i][j], end=" ")
        print()
```

```python
AgentPosition(environment,Agent)
```
